```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Some examples and explanation for plotting graphs

## Scatter Plot

```python
# the below plot is called as scatter plot and it takes 2 values i.e, of x and y
x=[1,2,3]
y=[2,5,6]
plt.scatter(x,y)
plt.ylabel("y")
plt.xlabel("x")
plt.title("Scatter Plot")
plt.show()
```

## Simple Line Plot

```python
# This is a simple line plot and takes two values of x and y
x=[1,2,3,4,5]
y=[2,5,3,2,6]
plt.plot(x,y)
plt.ylabel("y")
plt.xlabel("x")
plt.title("Simple Line plot")
plt.show()
```

## Plot function x^3

```python
# this is a function of x^3 plotted in graph using numpy
import numpy as np
#x=np.array([1,2,3,4])
x=np.arange(0,5,0.1)
# to make the line smooth
y=x**3
plt.plot(x,y,"bo") # 'bo' here is used for the dotted line plotted in the graph below 
plt.title("y = x^3")
plt.show()
```

## Plotting with categorical variables

#### It is also possible to create a plot using categorical variables. Matplotlib allows you to pass categorical variables directly to many plotting functions. For example:

```python
names = ['Salmanbhai', 'Aswathamma', 'GaneshGaitonde']
values = [1, 10, 100]

# this write here defines the size of the plotted graph, can be changed accordingly
plt.figure(figsize=(9,3))

plt.subplot(131)
plt.bar(names, values)
plt.xticks(rotation=90) # this is the method used to rotate the labelling on the x-axis by an angle of 90 degree in this case.

plt.subplot(132)
plt.scatter(names, values)
plt.xticks(rotation=90)

plt.subplot(133)
plt.plot(names, values)
plt.xticks(rotation=90)

plt.suptitle('Categorical Plotting')
plt.show()
```

## Histogram

```python
# random.random_integers creates the list for any 5 random integer values between 1-100  
x = np.random.random_integers(1, 100, 5)

plt.figure()

plt.hist(x)
plt.ylabel('No. of times')
plt.show()
```
